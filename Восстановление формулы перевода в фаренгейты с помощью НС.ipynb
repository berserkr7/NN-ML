


Восстановим формулу перевода в Фаренгейты F = C*1.8 + 32

```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras.layers import Dense
```

```python
# Обучающая выборка

C = np.array([-40, -10, 0, 8, 15, 22, 38])
F = np.array([-40, 14, 32, 46, 59, 72, 100])
```

```python
# Sequential() создает модель многослойной НС
# Класс Dense создает слой нейронов в полносвязной НС
model = keras.Sequential()
model.add(Dense(units=1, input_shape=(1,), activation='linear'))
```

```python
# Теперь зададим функцию потерь и надстройку ГС
model.compile(loss='mean_squared_error', optimizer=keras.optimizers.Adam(0.1)) # 0.1 это шаг сходимости ГС
```

```python
model
```

```python
history = model.fit(C, F, epochs=500, verbose=False)

plt.plot(history.history['loss'])
plt.grid(True)
plt.show()
```

```python
model.predict([100])
```
