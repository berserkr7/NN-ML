```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 5 # количество точек в каждом классе
b = 3 # смещение

x1 = np.random.random(N) # генерирует числа от 0 до 1 N штук
x2 = x1 + [np.random.randint(10)/10 for i in range(N)] + b
```

```python
C1 = [x1, x2]
```

```python
x1 = np.random.random(N) # генерирует числа от 0 до 1 N штук
x2 = x1 - [np.random.randint(10)/10 for i in range(N)] - 0.1 + b
```

```python
C2 = [x1, x2]
```

```python
f = [0 + b, 1 + b]
```

```python
f
```

```python
w2 = 0.5
w1 = -0.5
w3 = -b*w2

w = np.array([w1, w2, w3])
```

```python
for i in range(N):
    x = np.array([C2[0][i], C2[1][i], 1])
    y = np.dot(w, x)
    if y >= 0:
        print("Класс С1")
    else:
        print("Класс С2")
        
```

```python
plt.scatter(C1[0][:], C1[1][:], s=10, c='red')
plt.scatter(C2[0][:], C2[1][:], s=10, c='blue')
plt.plot(f)
plt.grid(True)
```
