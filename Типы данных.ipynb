# Кортежи

```python
#Кортеж также является последовательность. Кортеж неизменяем, то есть добавляем в него некоторый элемент и большего его изменить нельзя!
# Кортежи поддерживают теже операции, что и списки за исключением, тех, что изменяют содержание!!!!!!!!!!!
```

```python
# Создание кортежа
empty_tuple = () # создали пустой кортеж
# Чтобы создать кортеж, один элемент, нужно после элемента поставить запятую:
new_tuple = 'Alex', # кортеж из одного элемента
# Если в кортеже более одного элемента, то в конце запятую не ставим:
new_tuple = ('Alex', 'Derden')
# Создание кортежа из другого типа данных:
new_tuple = tuple(['hello', 'alo', 45, 23])
print(new_tuple)
new_tuple1 = tuple('dqwdqwd') # из строки сделали кортеж из элементов этой строки 
print(new_tuple1)
```

```python
# Кортежи позволяют присваивать значения нескольким переменным одновременно:
mark_tuple = ('One', 'Two', 5)
a,b,c = mark_tuple
print(a,b,c)
```

```python
# Также с помощью кортежа можно делать  swap значений:
a,b = b,a
```

```python
# Объединяем кортежи с помощью оператора + (это похоже на объединение строк)
print(('hello', 23, 34) + (1,2,3,4,5,6))
```

```python
# Размножаем элементы с помощью оператора * (принцип похож на многократное использование +)
print(('hello', 23, 34)*6)
print(('alex',)*5)
```

```python
# Сравнение кортежей
a = (1, 2, 3)
b = (1, 2, 3, 4)
print(a==b)
print(a<=b)
print(b>a)
```

```python
# Поскольку кортеж это последовательнось, то по нему можно итерировать с помощью for и in
words = (1,2,3,4,5,6)
for num in words:
    print(num)
```

```python
# Изменяем кортеж! Нееет, это невозможно. Но можно использовать как и в строках конкатенацию
a = (1, 2, 3)
b = (1, 2, 3, 4)
print(a)
a = a + b # это не тот же кортеж, у него уже другой id()!!!!
print(a)
```

```python
#Сортировка и реверс кортежа:
a = (1, 2, 3, 3, 5, 3, 9, 8)
# Это можно сделать только так!!!1
new_tuple = tuple(sorted(a))
print(new_tuple)
new_tup = tuple(reversed(a))
print(new_tup)
```

# Словари

```python
# Словарь очень похож на список, но порядок элементов в нем не имеет значения, значения выбираются по ключу. 
# Cловаь относится к изменяемым типам данных.Внимани! Ключ словаря обязан быть неизменяемым типом!!!! Значение - мб любым!
# Словари также называются ассоциативными массивами, хэш-таблицами, хэшами.
```

```python
# Создаем словарь 
empty_dict = {}
new_empty = dict()
new_dict = {1:'one', 2:'two', 3:'three'}
print(new_dict)
print(*new_dict) # при распаковке выводятся только ключи словаря
# Еще один способ создания
slovar = dict(one=1, two=2) #имена ключей должны быть уникальные имена переменных!!!!!!!
print(slovar)
```

```python
# Преобразуем с помощью dict()
# То есть словари можно создать и из других структур данных
lol = [[1,'one'], [2,'two'], [3,'three'], [4,'four']]
print(dict(lol))
# То есть можем использовать любую последовательность, содержащую последовательность из 2 элементов
lol = ([1,'one'], [2,'two'], [3,'three'], [4,'four'])
print(dict(lol))
lol = [(1,'one'), [2,'two'], [3,'three'], [4,'four']]
print(dict(lol))
lol = ('ab', 'cd', 'ef')
print(dict(lol))
lol = ['ab', 'cd', 'ef']
print(dict(lol))
```

```python
# Добавляем или изменяем элемент с помощью конструкции[ключ]. Добавить элемент в словарь легко:
# надо просто обратиться к элементу по его ключу и присвоить значение. Если ключ, существует, то имеющееся значение будет 
# заменено новым, если нет такого, то создаст новую пару и добавит в словарь.
new_dict = {
    'name':'ALEX',
    'last_name':'Derden',
    'age':28
}
print(new_dict)
new_dict['name'] = 'Ramazan'
new_dict['sex'] = 'men'
print(new_dict)
```

```python
# Получаем элемент словаря с помощью конструкции [ключ] или метода get()
new_dict = {
    'name':'ALEX',
    'last_name':'Derden',
    'age':28
}
print(new_dict['name'])
# Но если такого ключа в словаре нет, то будет сгенерировано исколючение!!!!!!
# Есть 2 способа избежать исключения: 

#1) 1 способ с помощью ключевого слова in, проверим, есть ли такой ключ:
'name' in new_dict

#2) 2 cпособ - использовать специальную функцию словаря - get(). Мы указываем словарь, ключ, и опциальное значение.
# Если ключ существует, то получим связанное с ним значение, если ключа нет, то получим опциональное значение
print(new_dict.get('name', 'net takogo'))
print(new_dict.get('jeneric', 'net takogo'))
print(new_dict)

# Если у get() не указать опциональный параметр, то по умолчанию там будет находиться None
```

```python
# Получаем все ключи с помощью функции keys(), ключи будут находитсья в итерабельном списке dict_keys() и чтобы 
# использовать методы списка - сначала надо преобразовать его к списку с помощью dict()
new_dict = {
    'name':'ALEX',
    'last_name':'Derden',
    'age':28
}
print(new_dict.keys(), type(new_dict.keys()))

#  Получаем все значения с помощью функции values(), она возращает также не простой список!!!
print(new_dict.values(), type(new_dict.values()))


# Получаем все пары ключ-значение с помощью  функции items(), где каждая пара будет возвращена в виде списка кортежей
print(new_dict.items(), type(new_dict.items()))
print(*new_dict.items())

```

```python
# C помощью функции len() - можем получить количество пар в словаре
new_dict = {
    'name':'ALEX',
    'last_name':'Derden',
    'age':28
}

print(len(new_dict))
```

```python
# Объединяем словари с помощью конструкции {**a, **b}
square = {i**2:i for i in range(10)}
cube = {i:i for i in range(10)}
c = {**square, **cube}
print(square)
print(cube)
print(c)
# Фактически конструкция {**a, **b, ...} может принимать более двух ключей
```

```python
# Объединяем словари с помощью функции update() - копируются ключи и значения в другой словарь, если в другом словаре
# есть такие же ключи, то победят значения из второго словаря
square = {i**2:i for i in range(10)}
cube = {i:i**3 for i in range(10)}
square.update(cube)
print(square)
```

```python
# Удаляем элементы по их ключу с помощью оператора del
new_dict = {
    'name':'ALEX',
    'last_name':'Derden',
    'age':28
}
print(new_dict)

del new_dict['age']
print(new_dict)
```

```python
# Получаем элемент по ключу и удаляем его с помощью функции pop()
# В этой функции объединены функции get() и del. Если в качестве аргумента передать ключ и такой ключ имеется, то 
# вернется соответствующее значение и пара удалится. Если ключа нет, то БУДЕТ СГЕНЕРИРОВАНО ИСКЛЮЧЕНИЕ!

new_dict = {
    'name':'ALEX',
    'last_name':'Derden',
    'age':28
}
print(new_dict.pop('age'))
print(new_dict)
```

```python
# Удаляем все элементы с помощью функции clear()
# Чтобы удалить все ключи и значения из словаря - достаточно использовать функцию clear() или же переназначить заданному имени пустой словарь.

new_dict = {
    'name':'ALEX',
    'last_name':'Derden',
    'age':28
}

print(new_dict)
new_dict.clear()
print(new_dict)

new_dict = {
    'name':'ALEX',
    'last_name':'Derden',
    'age':28
}
print(new_dict)
new_dict = {} # так тоже можно)
print(new_dict)
```

```python
# Проверяем на наличие ключа с помощью оператора in

new_dict = {
    'name':'ALEX',
    'last_name':'Derden',
    'age':28
}

print('age' in new_dict)
```

```python
# Присваиваем значения с помощью оператора = 
# Как в случае и со списками, если мы изменим словарь, то изменятся и все словари на которые ОН ССЫЛАЕТСЯ!!

new_dict = {
    'name':'ALEX',
    'last_name':'Derden',
    'age':28
}

copy_dict = new_dict

print(new_dict)
print(copy_dict)

new_dict['age'] = 2000

print(new_dict)
print(copy_dict)
```

```python
# Копируем все значения с помощью copy()
# Проблема выше решается с помощью copy()

new_dict = {
    'name':'ALEX',
    'last_name':'Derden',
    'age':28
}

copy_dict = new_dict.copy()

print(new_dict)
print(copy_dict)

new_dict['age'] = 2000

print(new_dict)
print(copy_dict)

# Но эта копия является поверхностной. Ее можно применять тогда, когда значения словаря неизменяемы. Если же изменяемы,
# то необходимо использовать deepcopy() 
```

```python
# Для получения пар-ключ значений с итерацией, воспользуемся items()

new_dict = {
    'name':'ALEX',
    'last_name':'Derden',
    'age':28
}

for item in new_dict.items():
    print(item, type(item))
```

```python
# Включение словарей 
# {выражение для ключа : выражение для значения for выражение in iterable }
print({i**2 : i for i in range(10)})

new_dict = {
    'name':'ALEX',
    'last_name':'Derden',
    'age':28
}

print({item[0] : item[1] for item in new_dict.items()})

word = 'letters' # подсчет встречаемости буквы
letter_dict_count = {letter : word.count(letter) for letter in set(word)}
print(letter_dict_count)
```

# Множества

```python
# Множества похоже на словарь, в котором значения отброшены, а оставлены только ключи. Ключи должны быть также уникальны.
# Множество - изменяемый, неупорядоченный тип данных.
```

```python
# Cоздаем множество с помощью функции set()
empty_set = set()
new_set = {1,2,3,4, 5, 5, 5, 5, 5, 5}
print(new_set)
# Порядок ключей в множестве не имеет значение!
# Также можем преобразовывать и другие данные
text = 'Hello, Boys!'
print(set(text))

new_list = [1, 2, 2, 3, 4, 'hello']
print(set(new_list))

# Если передать функции set() словарь, то она вернет ТОЛЬКО КЛЮЧИ!

new_dict = {
    'name':'ALEX',
    'last_name':'Derden',
    'age':28
}

print(set(new_dict))
print(*set(new_dict))
```

```python
# Получаем длину множества с помощью функции len()
print(len({1,2,2,2,2,2,2,2,2,2,5}))
```

```python
# Добавляем элемент в множество с помощью фунции add()
new_set = {1,2,3,4, 5, 5, 5, 5, 5, 5}
print(new_set)

new_set.add('hello')
print(new_set)
```

```python
# Удаляем элемент с помощью функции remove()
new_set = {1,2,3,4, 5, 5, 5, 5, 5, 5}
new_set.remove(5)
print(new_set)
new_set.remove('hello')
print(new_set)
```

```python
# Комбинации и операторы

# Объединение | или union()
# Пересечение & или intersection()
# Разность - или difference()
# Симметрическая разность ^ или symmetric_difference()

# В своем изначальном виде эти функции множеств нужно сохранять в новое множество результат их операции
a = {1, 2, 3, 4, 5, 6, 7, 8, 9}
b = {1, 2, 3, 4, 5, 20, 30, 40, 'Hello'}
print(a|b)
print(a&b)
print(a-b)
print(a^b)

# Если хотим изменить изначальное множество, то надо использовать update(), 
```
