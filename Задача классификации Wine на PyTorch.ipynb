```python
x_train, x_test, y_train, y_test = train_test_split(
    wine.data[:, :2],
    wine.target,
    test_size=0.3,
    shuffle=0.3

)
```

```python
# занесем данные в тензоры PyTorch

x_train = torch.FloatTensor(x_train)
x_test = torch.FloatTensor(x_test)
y_test = torch.LongTensor(y_test)
y_train = torch.LongTensor(y_train)

```

# Модель

```python
class WineNet(torch.nn.Module):
    def __init__(self, n_hidden_neurons):
        super(WineNet, self).__init__()
        
        self.fc1 = torch.nn.Linear(2, n_hidden_neurons)
        self.act1 = torch.nn.Sigmoid()
        self.fc2 = torch.nn.Linear(n_hidden_neurons, n_hidden_neurons)
        self.act2 = torch.nn.Sigmoid()
        self.fc3 = torch.nn.Linear(n_hidden_neurons, 3)
        self.sm = torch.nn.Softmax(dim=1)
        
    
    
    def forward(self, x):
        x = self.fc1(x)
        x = self.act1(x)
        x = self.fc2(x)
        x = self.act2(x)
        x = self.fc3(x)
        return x
        
    def probability_softmax(self, x):
        x = self.forward(x)
        x = self.sm(x)
        return x
    
wine_net = WineNet(5)
```

```python
# фукция потерь CrossEntropyLoss используется для тех данных, что еще не прошли
# через softmax

loss = torch.nn.CrossEntropyLoss()

optimizer = torch.optim.Adam(wine_net.parameters(), lr=0.001)
```

```python
np.random.permutation(5) # перестановка
```

```python
len(x_train)
```

```python
batch_size = 10

for epoch in range(10000):
    order = np.random.permutation(len(x_train))
    for start_index in range(0, len(x_train), batch_size):
        optimizer.zero_grad()
        
        batch_indexes = order[start_index : start_index + batch_size]
        
        x_batch = x_train[batch_indexes]
        y_batch = y_train[batch_indexes]
        
        preds = wine_net.forward(x_batch)
        
        loss_val = loss(preds, y_batch)
        loss_val.backward()
        
        optimizer.step()
        
    if epoch % 100 == 0: # каждые 100 эпох проверяем обучается ли модель
        test_preds = wine_net.forward(x_test).argmax(dim=1)
        print((test_preds == y_test).float().mean())
```
