```python
def devirate(x1,x2):
    df1 = 2*x1
    df2 = 20*x2
    return np.array([df1,df2])

def gradient(position, lr, size=10):
    for i in range(size):
        res = np.array(position) - lr * devirate(position[0], position[1])
        print(f'a{i+1} = {list(position)} - {lr} * {list(devirate(position[0], position[1]))} = {list(res)}')
        position = res
        
gradient([10,1], 0.1, size=2)
```

```python
import sympy as sp

def diffx(x):
  return 2 * x
def diffy(y):
  return 20 * y

a = [10.0, 1.0]
h = 5
G = [1.0, 1.0]

for n in range(1,10):
    G[0] += diffx(a[0]) ** 2
    G[1] += diffx(a[1]) ** 2
    a[0] -= h / sp.sqrt(G[0]) * diffx(a[0])
    a[1] -= h / sp.sqrt(G[1]) * diffy(a[1])
    print('G[', n, ']=[', G[0], ',', G[1], ']', sep='')
    print('a[', n, ']=[', a[0], ',', a[1], ']', sep='')
```

```python
import sympy
import numpy as np

w0, w1 = sympy.symbols('w0 w1')
# sympy.diff позволяет получить производную в аналитическом виде
terms_der = [
    (sympy.diff((w0 - 1) ** 2, w0), (sympy.diff((w0 - 1) ** 2, w1))),
    (sympy.diff((w1 + w0 - 2) ** 2, w0), (sympy.diff((w1 + w0 - 2) ** 2, w1))),
    (sympy.diff((2 * w1 + w0 - 3) ** 2, w0), (sympy.diff((2 * w1 + w0 - 3) ** 2, w1)))
]
epochs = 25
pos = np.array([0, 0])
h = 0.1
# sympy.lambdify позволяет использовать аналитическую производную с конкретными значениями
for i in range(epochs):
    order = np.random.choice(np.arange(len(terms_der)), len(terms_der), replace=False)
    print(f'Эпоха номер {i + 1}:')
    for j in range(len(order)):
        grad = [
            sympy.lambdify((w0, w1), terms_der[order[j]][0])(pos[0], pos[1]),
            sympy.lambdify((w0, w1), terms_der[order[j]][1])(pos[0], pos[1])
        ]
        pos = pos - h * np.array(grad)
        print(f'Итерация номер {j + 1}, новая позиция - {pos[0]}, {pos[1]}')
```

```python
def deriv_mod(x):
    return 1 if x > 0 else (0 if x == 0 else -1)

derivative = [lambda x: 6 * x[0] + deriv_mod(x[0]),
              lambda x: 4 * x[1] + 4 + deriv_mod(x[1])]  #Производные по w0 и по w1
a = [0, 0]
number_of_iterations = 2 #Число шагов
h = 0.1
for n in range(number_of_iterations):
    w0 = a[0]
    w1 = a[1]
    a[0] -= h * derivative[0]([w0, w1])
    a[1] -= h * derivative[1]([w0, w1])
    print('Шаг: {} w0 = {} w1 = {}'.format(str(1 + n).ljust(8), str(round(a[0], 3)).ljust(8), str(round(a[1], 3)).ljust(8)))
```
